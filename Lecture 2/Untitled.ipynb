```python
import numpy as np
```

```python
def trials(m,p):
    x=np.random.uniform(0,1,m)
    k=len(x[x > p])
    y=np.random.uniform(0,1)
    S=(k+1)/(m+2)
    return [S, y]
    
    
```

```python
N=100
p=0.1
m=30
S=np.zeros(N)
y=np.zeros(N)
for i in range(N):
    S[i], y[i] = trials(m,p)
    
```
